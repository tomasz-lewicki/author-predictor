```python
import pandas as pd
import numpy as np
```

```python
train, test = pd.read_csv('./data/train.csv'), pd.read_csv('./data/test.csv')
```

# Create author name -> int mapping

```python
train
```

```python
train.author.values
```

```python
from collections import OrderedDict
label_meaning = OrderedDict()
for idx, l in enumerate(train.author.unique()):
    label_meaning[l] = idx
    
print(label_meaning)
```

```python
y = np.array(list(map(label_meaning.get, train.author.values)))
y
```

Now we have our labels in a desirable format

```python
X_raw = train.text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(min_df=3,  max_features=None,
                             strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
                             ngram_range=(1, 3), use_idf=1,smooth_idf=1,sublinear_tf=1,
                             stop_words = 'english')

vectorizer.fit(X_raw)
X_tfidf =  vectorizer.transform(X_raw) 
```
